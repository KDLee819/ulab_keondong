```python
# Your code here

# Create a figure
plt.figure(figsize=(5, 5))

# Define indep. and dep. variables
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Plot the data
# First plot: line plot
plt.subplot(1, 2, 1)
plt.plot(x, y, color="red", linestyle="--", alpha=0.2) # feature 'alpha'
plt.title(r"Line plot of $y=2x$", fontsize="16")
plt.xlabel("values of x", fontsize="12")
plt.ylabel("values of y", fontsize="12")
plt.legend("Line plot")

# Second plot: scatter plot
plt.subplot(1, 2, 2)
plt.scatter(x, y, color="green", marker="o", alpha=1) # affects transparency of the plot
plt.title(r"Scatter plot of $y=2x$", fontsize="16")
plt.xlabel("values of x", fontsize="12")
plt.ylabel("values of y", fontsize="12")
plt.legend("Scatter plot")

# Added because two graphs would squeezed too much without it
plt.tight_layout() 
plt.show()
```

# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Your code here

def plot_square_and_sqrt(xlist):
    '''
    This is the function that plots two graphs of x vs y
    One is f(x)=x^2 and another is g(x) = sqrt(x)
    It will return two functions in a same plot, and lists of f and g values

    Input: a list of x massaged by function np.linspace
    Output: a plot of two functions, and a tuple of lists containing their values
    '''
    # Make an empty list for values of f(x) and g(x) to be stored
    f_of_x = []
    g_of_x = []

    # Iterate for x-values in the xlist
    for x_values in xlist:
        f_of_x.append(x_values**2)       # f(x) = x^2
        g_of_x.append(np.sqrt(x_values)) # g(x) = sqrt(x)

    # After generating two lists of f and g, create a plot with necessary attributes
    plt.figure(figsize=(10,8))
    plt.plot(xlist, f_of_x, label=r"$f(x)=x^2$", color="orange", linestyle="--")
    plt.plot(xlist, g_of_x, label=r"$g(x)=\sqrt{x}$", color="purple", linestyle=":")
    
    plt.xlabel("values of x")
    plt.ylabel("values of y")
    plt.title(r"Two functions $f(x)=x^2$ and $g(x)=\sqrt{x}$ for given domain of x")
    
    plt.legend()
    plt.show()

    # Also, don't forget to return the list of f(x) and g(x) values!
    return f_of_x, g_of_x

# This is our xlist:
xlist = np.linspace(0, 10, 100)

plot_square_and_sqrt(xlist)
```

Make a **new function** (that has the same foundation as your first function in this question) but includes **all** the changes you described in your comments. Don't forget to call the function and show its output. 

```python
# Your code here

def plot_square_and_sqrt_with_sum(xlist):
    '''
    This is the function that plots two graphs of x vs y
    One is f(x)=x^2 and another is g(x) = sqrt(x)
    Additionally, it plots the sum of f(x) and g(x) as well
    It will return three functions in a same plot, and lists of f, g, f+g values

    Input: a list of x massaged by function np.linspace
    Output: a plot of two functions, plot of their sum,
            and a tuple of three lists containing their values
    '''
    # Make an empty list for values of f(x) and g(x) to be stored
    f_of_x = []
    g_of_x = []

    # Iterate for x-values in the xlist
    for x_values in xlist:
        f_of_x.append(x_values**2)       # f(x) = x^2
        g_of_x.append(np.sqrt(x_values)) # g(x) = sqrt(x)

    # Make a list of the sum of f and g
    f_g_sum = list(np.array(f_of_x) + np.array(g_of_x))

    # After generating the three lists, create a plot with necessary attributes
    plt.figure(figsize=(10,8))
    plt.plot(xlist, f_of_x, label=r"$f(x)=x^2$", color="orange", linestyle="--")
    plt.plot(xlist, g_of_x, label=r"$g(x)=\sqrt{x}$", color="purple", linestyle=":")
    # New function of f+g
    plt.plot(xlist, f_g_sum, label=r"$f(x)+g(x)$", color="green", linestyle="-")

    # Vertical dashed line at x=13
    plt.axvline(x=13, color="red", linestyle='--', label='Vertical Line')
    
    # Horizontal dotted line at y=100
    plt.axhline(y=100, color="blue", linestyle=':', label='Horizontal Line')
    
    plt.xlabel("values of x")
    plt.ylabel("values of y")
    plt.title(r"Three functions: $f(x)=x^2$, $g(x)=\sqrt{x}$ and $f(x)+g(x)$ for given domain of x")

    plt.xlim(4, 16)  # Newly added x-limit
    plt.ylim(0, 200) # Newly added y-limit
    plt.legend(loc="upper right", bbox_to_anchor=(1.22, 1)) # Newly added legend attributes
    plt.show()

    # Also, don't forget to return the list of f(x), g(x) and f+g values!
    return f_of_x, g_of_x, f_g_sum

# Make our NEW xlist, shift by 5:
xlist = np.linspace(0, 10, 100)
new_xlist = list(np.array(xlist)+5)

plot_square_and_sqrt_with_sum(new_xlist)
```
